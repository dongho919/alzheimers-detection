```python
import numpy as np
import cv2 as cv
import os
import matplotlib.pyplot as plt

#load images
image1 = cv.imread('../Project/Alzheimer_s Dataset_binary/train/NonDemented/nonDem4.jpg')
image2 = cv.imread('../Project/Alzheimer_s Dataset_binary/train/NonDemented/nonDem1.jpg')
```

```python
images = []
labels = []
for root, dirs, files in os.walk('../Project/Alzheimer_s Dataset_binary/train/NonDemented/'):
    for name in files:
        filename = os.path.join(root, name)
        images.append(cv.imread(filename, cv.IMREAD_GRAYSCALE))
        labels.append("nonDem")
for root, dirs, files in os.walk('../Project/Alzheimer_s Dataset_binary/train/Demented/'):
    for name in files:
        filename = os.path.join(root, name)
        label = []
        for char in name :
            if not char.isnumeric():
                label.append(char)
            else:
                break
        labels.append("".join(label))
        images.append(cv.imread(filename, cv.IMREAD_GRAYSCALE))
```

```python
print(np.array(images).shape)
#print(images[0])
#plt.imshow(images[436])
#print(labels)
print(np.array(labels).shape)
print(labels)
```

```python
#convert to grayscale image
gray_scale1 = cv.cvtColor(image1, cv.COLOR_BGR2GRAY)
gray_scale2 = cv.cvtColor(image2, cv.COLOR_BGR2GRAY)
```

```python
#initialize SIFT object
sift = cv.xfeatures2d.SIFT_create()

keypoints1, des1= sift.detectAndCompute(image1, None)
keypoints2, des2= sift.detectAndCompute(image2, None)
```

```python
# initialize Brute force matching
bf = cv.BFMatcher(cv.NORM_L1, crossCheck=True)
matches = bf.match(des1,des2,)

```

```python
 #sort the matches 
matches = sorted(matches, key= lambda match : match.distance)

```

```python
#draw matches
matched_imge = cv.drawMatches(image1, keypoints1, image2, keypoints2, matches[:30], None)
cv.imshow("Matching Images", matched_imge)
cv.waitKey(0)

array1 = np.array([kp.pt for kp in keypoints1])
array2 = np.array([kp.pt for kp in keypoints2])

dists = [match.distance for match in matches[:30]]
np.mean(dists)
```
